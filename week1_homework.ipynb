```python
Rolls_Royce = df[df['Make'] == 'Rolls-Royce']
Rolls_Royce
```

```python
RR_3_feat=Rolls_Royce[["Engine HP","Engine Cylinders", "highway MPG" ]]
RR_3_feat
```

```python
RR_3_feat=RR_3_feat.drop_duplicates()
RR_3_feat
```

```python
X = RR_3_feat.values
X
```

```python
XT = X.T
XT
```

```python
XTX = XT.dot(X)
XTX
```

```python
XTX_INV = np.linalg.inv(XTX)
XTX_INV
```

```python
#Sum of elements of the inverse matrix
XTX_INV.sum()
```
