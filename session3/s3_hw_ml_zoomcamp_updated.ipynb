```python
#Calculating Correlation between numerical features
def corre():

    numerical = ['latitude','longitude','minimum_nights','number_of_reviews',
    'reviews_per_month','calculated_host_listings_count','availability_365']

    cor_list = []
    cor_feat = []

    for i in numerical:
        (df_train[numerical].corrwith(df_train[numerical][i]))
        cor_feat.append("Correlated_with %s" %i)
        cor_list.append((df_train[numerical].corrwith(df_train[numerical][i])))
        


    return cor_feat,cor_list
```

```python
#removing each feature individually and storing in new data frames
def feat_remove():

    features = categorical + numerical

    feat_train = []
    feat_val = []
    feat_i = []
    
    for i in features:
      feat_train.append(df1_train.drop([i],axis=1))
      feat_val.append(df1_val.drop([i],axis=1))
      feat_i.append(i)

    return feat_train,feat_val,feat_i
```

```python
#data frames with individual features being removed to train and compute the model
feat_train,feat_val,feat_removed = feat_remove()
```

```python
#function for storing the accuracy values using feature elimination technique 
def train_model(ft,fv,fr):
    

    from sklearn.feature_extraction import DictVectorizer


    dv = DictVectorizer(sparse=False)

    train_dict = ft.to_dict(orient='records')
    X_train = dv.fit_transform(train_dict)

    val_dict = fv.to_dict(orient='records')
    X_val = dv.transform(val_dict)

    from sklearn.linear_model import LogisticRegression

    model = LogisticRegression(solver='liblinear', random_state=42)
    
    model.fit(X_train, above_average)
    


    
    y_pred_binar = model.predict(X_val)
    
    
    price_val_binary = (y_val>=152).astype('int')

   
    accuracy_y_pred_val = (price_val_binary==y_pred_binar).mean()
    
    return accuracy_y_pred_val,fr
```

```python

#function for storing the accuracy values using feature elimination technique in accuracy_val = [] list
def feat_remov_accuracy(ft,fv,fr):

    accuracy_val = []

    for i in range(9):
        accuracy_val.append(train_model(ft[i],fv[i],fr[i]))
        

    return accuracy_val
```

```python
#storing the calculated accuracy value differ
accuracy_values = feat_remov_accuracy(feat_train,feat_val,feat_removed)
```

```python
#Feature selection - the smallest difference in accuracy
accuracy_values
```

```python
#For loop calculating and storing the difference between original accuracy and accuracy calculated with feature elimination technique
accuracy_dif = []
removed_feature = []
orig_accuracy

for i in range(9):
        diff = orig_accuracy - accuracy_values[i][0]
        accuracy_dif.append(diff)
        removed_feature.append(accuracy_values[i][1])

        
```

```python
#accuracy difference and corrsponding feature being removed
print("The accuracy difference values with feature elimination technique are : {} \nThe removed features of the corresponding accuracy difference values are : {}\n".format(accuracy_dif,removed_feature))

print("The smallest difference out of all the feature is : {} \nThe corresponding feature being removed is: {} \n\n ".format(min(accuracy_dif),removed_feature[np.where(accuracy_dif == min(accuracy_dif))[0][0]]))

print("The smallest different out of "'neighbourhood_group'"," 'room_type'","'number_of_reviews'"," 'reviews_per_month'" is : {} \nThe corresponding removed feature is :{}".format(accuracy_dif[5],removed_feature[5]))
```

```python
#Applying the logarithmic transformation to target variable price for all sets
y_train_log = np.log1p(y_train)
y_val_log = np.log1p(y_val)
y_test_log = np.log1p(y_test)
```

```python
y_train_log
```

```python
#function for calculating rmse
def rmse(y,y_pred):
    se = (y - y_pred)**2
    mse = se.mean()
    return np.sqrt(mse)
```

```python
#function for calculating prediction values for different alpha parameters in Ridge regression model
def train_ridge_model(tr,val,alpha):
    

    from sklearn.feature_extraction import DictVectorizer


    dv = DictVectorizer(sparse=False)

    train_dict = tr[categorical + numerical].to_dict(orient='records')
    X_train = dv.fit_transform(train_dict)

    val_dict = val[categorical + numerical].to_dict(orient='records')
    X_val = dv.transform(val_dict)

    from sklearn.linear_model import Ridge

    model = Ridge(alpha = alpha)
    
    model.fit(X_train, y_train_log)
    


    
    y_pred = model.predict(X_val)
    
    
    price_val_binary = (y_val_log>=152).astype('int')

   
    accuracy_y_pred_val = (price_val_binary==y_pred).mean()
    
    return y_pred,alpha
```

```python
#For loop storing prediction values for different alpha parameters in Ridge regression model in y_pred_values = []
y_pred_values = []
alpha_values  = []

for i in [0,0.01,0.1,1,10]:
    y_pred,alpha = train_ridge_model(df_train,df_val,i)
    y_pred_values.append(y_pred)
    alpha_values.append(i)
    
```

```python
#Ridge regression model prediction values and corresponding alpha values

print("The prediction values for Ridge regression model are : {} \n \nThe corresponding Alpha Parameters are : {}".format(y_pred_values,alpha_values))

```
